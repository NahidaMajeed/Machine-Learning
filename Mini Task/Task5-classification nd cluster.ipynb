```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
```

```python
iris=load_iris()
iris
```

```python
x=iris.data
x
```

```python
y=iris.target
y
```

```python
# split the data into training and testing set
x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# train a logistic regression classifier
classifier=LogisticRegression(max_iter=1000)
classifier.fit(x_train,y_train)
```

```python
# evaluate the accuracy of the classifier
y_pred=classifier.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy of Logistic Regression Classifier:", accuracy)
```

```python
# use trained model to predict classes of new samples
new_samples = np.array([[5.1, 3.5, 1.4, 0.2],  # Septosa
                        [6.2, 2.9, 4.3, 1.3],  # Versicolour
                        [7.7, 3.0, 6.1, 2.3]])  # Virginica

predicted_classes=classifier.predict(new_samples)
for sample, predicted_classes in zip(new_samples, predicted_classes):
    print(f"Predicted classes of new samples: {predicted_classes}- {iris.target_names[predicted_classes]}")
```

```python
# perform clustering on the dataset using different algorithms
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_clusters = kmeans.fit_predict(x)

agg = AgglomerativeClustering(n_clusters=3)
agg_clusters = agg.fit_predict(x)

gmm = GaussianMixture(n_components=3, random_state=42)
gmm_clusters = gmm.fit_predict(x)

plt.figure(figsize =(12,4))

plt.subplot(1,3,1)
plt.scatter(x[:,0],x[:,1],c=kmeans_clusters,cmap="magma")
plt.title("Kmean Clustering")

plt.subplot(1,3,2)
plt.scatter(x[:,0],x[:,1],c=agg_clusters,cmap="magma")
plt.title("Agglomerative Clustering")

plt.subplot(1,3,3)
plt.scatter(x[:,0],x[:,1],c=gmm_clusters,cmap="magma")
plt.title("Gaussian Mixture Model Clustering")

plt.show()
```

```python
print(accuracy_score(y, kmeans_clusters))
print(accuracy_score(y, agg_clusters))
print(accuracy_score(y, gmm_clusters))
```
