## 8. Cleaning Name Field
Plan: 
1. Clear all the Null values
2. Remove Title in name if any
3. remove the extraneous words if any
****

```python
df = df.dropna(subset=['Name'])
```

```python
df
```

**handle name with titles**

```python
name_title_pattern = r'\b(?:Mr\.|Mrs\.|Ms\.|Dr\.|Miss\.)\s'
# Filter rows where names contain titles
names_with_titles = df[df['Name'].str.contains(name_title_pattern, na=False)]

print("Names with titles:")
print(names_with_titles)
```

```python
import re

# replace titles in a string
def replace_titles(name):
    if isinstance(name, str):  # Check if name is a string
        return re.sub(name_title_pattern, '', name)
    else:
        return name  # Return NaN or non-string values unchanged :Just in case

# Apply the replace_titles function to the 'Name' column
df.loc[:,'Name'] = df['Name'].apply(replace_titles)
```

```python
df[df['Name'] == "Sheri Duffyparty"]
```

```python
df[df['Name'].str.contains("Ann Heath", na=False)]
```

**handle 3rd name**

```python
# Function to remove the third word from names with length 3
def remove_third_word(name):
    if isinstance(name, str):  # Check if the value is a string
        words = name.split()  # Split the name into words
        if len(words) == 3:  # Check if the name has exactly 3 words
            del words[2]  # Remove the third word
            return ' '.join(words)  # Join the remaining words back together
    return name  # Return the name unchanged for non-string values or names with different lengths

# Apply the remove_third_word function to the 'Name' column
df.loc[:,'Name'] = df['Name'].apply(remove_third_word)
```

```python
last_names = df['Name'].str.split().str[-1]
```

```python
last_names[last_names.str.contains("Taylor", na=False)]
```

```python
# Get the counts of each name
name_counts = last_names.value_counts()
total_names = name_counts[name_counts > 0].index.tolist()

# Filter names that are repeated more than once
repeated_names = name_counts[name_counts > 2].index.tolist()
# Filter the DataFrame to include only rows with repeated names
valid_surname = last_names[last_names.isin(repeated_names)]
```

```python
valid_surname = list(set(valid_surname))
valid_surname = sorted(valid_surname, key=len, reverse=True)
valid_surname
```

**some more filtering needed:**

**checking the common in (total_names and valid_surname) to get the final surname list**

```python
#total_names = ['taylordaughter', 'taylorphysical', 'lambert', 'lambsoft', 'mooney']
#valid_surname = ['taylor', 'lambert', 'Hernandez', 'lamb']

# Initialize an empty list to store converted elements
converted_list = []

# Iterate through list1
for item in total_names:
    # Check if any valid surname is a substring of the item
    for surname in valid_surname:
        if surname in item:
            converted_list.append(surname)
            break  # Stop searching for other surnames once a match is found
    else:
        # If no match is found, keep the original item
        converted_list.append(item)
converted_list = list(set(converted_list))
```

```python
##### this is the final list of SURNAMES Yuhuuuuuuuu
converted_list = sorted(converted_list, key=len, reverse=True)
check_surname_in_list = [surname for surname in converted_list if "Mooney" in surname]
check_surname_in_list
```

**copy the name to a new column and trimming out the extraneous words from the surname**

```python
# Initialize an empty list to store the matched parts
matched_parts = []

# Iterate through the DataFrame
for index, row in df.iterrows():
    # Extract the last name from the 'name' column
    last_name = row['Name New'].split()[-1]
    
    # Iterate through the surname list
    for s in converted_list:
        # Check if the current surname is a substring of the last name
        if s in last_name:
            # Store the matched part
            matched_parts.append(s)
            # Replace the last name with the matched part
            df.at[index, 'Name New'] = row['Name New'].rsplit(' ', 1)[0] + ' ' + s
            # Exit the loop since we found a match
            break
```

```python
index_of_name_column = df.columns.get_loc('Name')
salary_column = df.pop('Name New')  # Remove 'Salary' column from DataFrame
df.insert(index_of_name_column, 'Name New', salary_column)  # Insert 'Salary' column at the desired position
```
